```python
import pandas as pd

df=pd.read_csv("india_yearly_full_release_long_format.csv")

states_to_keep = ['Haryana','Uttar Pradesh','Madhya Pradesh','Bihar']
```

```python
df_filtered = df[df['State'].isin(states_to_keep)].copy()
```

```python
print("--- Head of your new filtered data ---")
print(df_filtered.head())

print("\n--- Unique states in your new data (for confirmation) ---")
print(df_filtered['State'].unique())
```

```python
output_filename = 'four_states_data.csv'
df_filtered.to_csv(output_filename, index=False)

print(f"\nSuccessfully created '{output_filename}' with data for your 4 states.")
```
